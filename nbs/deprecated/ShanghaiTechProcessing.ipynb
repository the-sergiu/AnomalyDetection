```python
import os
import cv2
```

```python
# Specify the directory containing the videos
video_directory = "./datasets/shanghaitech/videos/"

# List of common video file extensions
video_extensions = ['.mp4', '.avi', '.mov', '.mkv', '.flv', '.wmv']

# Get a list of video files in the directory
video_files = [file for file in os.listdir(video_directory) if any(file.lower().endswith(ext) for ext in video_extensions)]

print(video_files)
```

```python
# Specify the directory containing the test videos
test_video_directory = "./datasets/shanghaitech/testing/frames"

# Get a list of video files in the directory
test_video_files = [f"{file}.avi" for file in os.listdir(test_video_directory) ]
test_video_dirs = [file for file in os.listdir(test_video_directory) ]

print(test_video_files)
```

```python
train_frames_path = "./datasets/shanghaitech/training/frames/"
```

```python
def format_number(num):
    return "{:03d}".format(num)
# Test the function
print(format_number(1))   # Output: 001
print(format_number(20))  # Output: 020
print(format_number(300)) # Output: 300
```

```python
for video_file in video_files:

    dir_name = video_file.split(".")[0]

    video_path = os.path.join(video_directory, video_file)

    frames_directory = os.path.join(train_frames_path, dir_name + "/")

    os.makedirs(frames_directory, exist_ok=True)

    print(dir_name, video_path, frames_directory)
    
    # Create a VideoCapture object
    cap = cv2.VideoCapture(video_path)
    
    # Check if video opened successfully
    if not cap.isOpened():
        print("Error: Could not open video.")
        exit()

    frame_count = 0

    # Read until video is completed
    while cap.isOpened():
        # Capture frame-by-frame
        ret, frame = cap.read()
    
        if ret:
            # Save the frame
            frame_filename = os.path.join(frames_directory, f"{format_number(frame_count)}.jpg")
            # print(frame_filename)
            cv2.imwrite(frame_filename, frame)
            frame_count += 1
        else:
            # Break the loop when there are no frames left
            break

    # Release the VideoCapture object
    cap.release()
    print(f"Frames extracted for video {dir_name}.avi: {frame_count}")
```
